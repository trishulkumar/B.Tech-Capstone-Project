```python
'''
       Logistic Regression
       KNN
       Random Forest 
       Nural Networks 
       XGBoost 
       naiyes baiyes 
       Decision Tree 
       Decision Tree #with ada boost 
       
'''

Lr=67
knn=79
rf=83
svm=77
nb=66
xgb=74
nn=85
dtA=95


scores=[Lr,knn,rf,svm,nb,xgb,nn,dtA]
algorithm=['Logistic Regression','KNN','RandomForest','SVM','NaivesBayes','XgBoost','ANN','Decision-Tree + Ada boost']
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from numpy.random import rand

c = ['#0CBABA', '#5579C6', '#B91372', '#FBB034', '#20bf55','#cc2b5e','#7F53AC','#F9484A']
plt.bar(algorithm,scores,color=c)
##cc2b5e #5579C6 #753a88 #0CBABA
plt.ylabel('Accuracy',fontsize=10,fontname='monospace')
plt.xlabel('Algorithms ')
plt.title('Model Scores',fontsize=18,fontname='monospace')
plt.xticks(rotation=90,fontsize=10,fontname='monospace')
plt.ylim(0,100)
plt.show()


#plt.xticks(xticks_pos, labels,  ha='right', rotation=55, fontsize=20, fontname='monospace')
```
